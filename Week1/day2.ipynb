```python
#load the api key
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
```

```python
#VALIDATE API KEY
if not api_key:
    print('API key is not valid, Use the valid OPENAI API key')
elif not api_key.startswith('sk-proj'):
    print('API key is not a valid name')
elif api_key.strip() != api_key:
    print('API key is having trailing spaces')
else:
    print("API key is valid")
```
