```python
def call_website(website):
    prompt = [{'role':'system','content':system_prompt},
              {'role':'user','content':user_prompt + website}]
    response=openai.chat.completions.create(model='gpt-4.1-mini',messages=prompt)
    return response.choices[0].message.content
```

We can control the amount of randomness in the response using the temperature parameter.
Default vaulue is 1
0 - almost entirely deterministic
2 - highly random

```python
prompt = "Life is like a box of chocolates"
message=[{'role':'user','content':prompt}]
response = openai.chat.completions.create(model="gpt-5-nano",messages=message,temperature=1.0)
response.choices[0].message.content
```

```python
# Step 1: Create your prompts

def summarize(content):
    system_prompt = """You are a helpful assistant that analyze the content and provide appropriate email subject line.
                    Also summarize the content with three concise points about the product. 
                    """
    user_prompt = f"here is the content about the prodcut + {content}"


# Step 2: Make the messages list

    messages = [{'role':'system','content':system_prompt},
             {'role':'user','content':user_prompt}]

# Step 3: Call OpenAI
    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages, max_completion_tokens=100,temperature=2)

# Step 4: print the result
    return response.choices[0].message.content
```
